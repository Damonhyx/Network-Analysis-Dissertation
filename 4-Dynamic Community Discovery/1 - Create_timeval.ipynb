```python
import pandas as pd
from datetime import datetime
```

```python
path = "/Users/damon/Desktop/Dessertion/1-Data Collection/questions_2022-1-3.csv"
```

```python
df = pd.read_csv(path)
```

```python
df = df.sort_values(by=['Creation_date'])
```

```python
df
```

```python
tags = []
time_tags = []
```

```python
couple_tags = []
time_couple = []
```

```python
for i in df.values:
    tagsss = i[0].split("|-|")[:-1]
    for t in range(len(tagsss)):
        tags.append(tagsss[t])
        time_tags.append(datetime.utcfromtimestamp(i[8]).strftime('%Y-%m-%d %H:%M:%S'))
        for j in tagsss[t+1:]:
            couple_tags.append((tagsss[t], j))  
            time_couple.append(datetime.utcfromtimestamp(i[8]).strftime('%Y-%m-%d %H:%M:%S'))
```

```python
new_date = []
month = []
for i in df["Creation_date"]:
    new_date.append(datetime.utcfromtimestamp(i).strftime('%Y-%m-%d %H:%M:%S'))
    month.append(datetime.utcfromtimestamp(i).strftime('%m'))
```

```python
new_df = pd.DataFrame({"Title": df["Title"], "Timeval": new_date, "Month": month})
new_df.to_csv("/Users/damon/Desktop/Dessertion/4-Dynamic Community Discovery/question_with_time.csv", index=False)
```

```python
source = []
target = []
for i in couple_tags:
    source.append(i[0])
    target.append(i[1])
```

```python
month = []
for i in time_tags:
    month.append(i[5:7])
```

```python
new_df = pd.DataFrame({"Id": tags, "Timeval": time_tags, "Month": month})
new_df.to_csv("/Users/damon/Desktop/Dessertion/4-Dynamic Community Discovery/tags_with_time.csv", index=False)
```

```python
month = []
for i in time_couple:
    month.append(i[5:7])
```

```python
new_df = pd.DataFrame({"Source": source,
                       "Target": target,
                      "Timeval": time_couple, "Month": month})
new_df.to_csv("/Users/damon/Desktop/Dessertion/4-Dynamic Community Discovery/tags_edges_with_time.csv", index=False)
```
